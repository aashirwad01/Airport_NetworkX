```python
import networkx as nx
```

```python
routes=pd.read_csv("https://raw.githubusercontent.com/jpatokal/openflights/master/data/routes.dat",names=['Airline','Airline_ID','S_airport','S_airport_id','D_airport',"D_airport_id",'codeshare','stops','equipment'],header=None)
airport=pd.read_csv('https://raw.githubusercontent.com/jpatokal/openflights/master/data/airports.dat',header=None,names=['Airport_ID','Name','City','Country','IATA','ICAO','Latitude','Longitude','Altitude','T_Z','DST','Tz_dataset_time_zone','Type','Source'])
```

```python
routes.drop(routes[routes['S_airport_id']=='\\N'].index , inplace=True)

routes.drop(routes[routes['D_airport_id']=='\\N'].index , inplace=True)
```

```python
r_df=pd.DataFrame(columns=['Airline','Airline_ID','S_airport','S_airport_id','D_airport',"D_airport_id",'codeshare','stops','equipment'])
```

```python
k=[]
j=0
for i in (routes.index).values:
    if(int(routes['S_airport_id'][i]) in  id.values and int(routes['D_airport_id'][i]) in  id.values):
        
        k.append(routes.iloc[i])
    
       
           
```

```python
len(k)
```

```python
routes_df=r_df.append(k, ignore_index=True)
```

```python
routes_df
```

```python
edges= [(row['S_airport'], row['D_airport']) for item, row in routes_df[['S_airport', 'D_airport']].iterrows()]
route_graph= nx.Graph()
route_graph.add_edges_from(edges)
```
